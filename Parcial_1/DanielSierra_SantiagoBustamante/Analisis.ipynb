```python
import numpy as np
import ROOT
from ROOT import gROOT
from array import array
```

```python
Medias=np.array([]) #Esta lista va a almacenar los tiempos de ejecucion medios hallados para cada valor de N, que es la dimensi√≥n de la matriz del juego
Maximos=np.array([]) #Esta lista va a almacenar el tiempo maximo de ejecucion hallados para cada valor de N
```

```python
Data=open('datos_5.dat','r') #Primero abrimos el archivo de datos

Histograma_5 = ROOT.TH1F( 'h1f', 'Tiempos para n = 5', 13, 0, 3) #Definimos el histograma

data_5 = np.array([]) #Creamos el arreglo de datos

for i in Data: #Aqui almacenamos los valores del archivo datos_5.dat en el arreglo data_5
  data_5=np.append(float(i),data_5)
  Histograma_5.Fill(float(i))

c1 = ROOT.TCanvas( 'c1', 'N = 5', 200, 10, 700, 700 ) #Abrimos un canvas para poder dibujar
c1.SetFillColor(20) #Le damos un poco de estetica al asunto
c1.SetBorderMode(1)
c1.SetBorderSize(5)
c1.SetGrid()

f1 = ROOT.TF1("f1", "gaus", 0, 300); #Definimos aqui la funcion gaussiana con que vamos a hacer el fit de todas los histogramas

x = Histograma_5.GetXaxis(); y = Histograma_5.GetYaxis()
x.SetTitle("Tiempo [s]"); y.SetTitle("Frecuencia")
Histograma_5.Fit("f1","R") #Hacemos el fit para Histograma_5
Histograma_5.Draw() #Y finalmente dibujamos

c1.Update() #y actualizamos el canvas
c1.Draw()

Medias=np.append(Medias,Histograma_5.GetMean()) #Los valores obtenidos para media y maximo son almacenados
Maximos=np.append(Maximos,max(data_5))
```

```python
c13 = ROOT.TCanvas( 'c13', 'Tiempo medio vs N', 200, 10, 700, 500)
c13.SetGrid()

sq = ROOT.TF1("sq", "[0]*x**2", 1, 60) #De estos datos hacemos un fit de la forma ax**2
sq.SetParameter(0, 1)

n = array('d', range(5,65,5))
gr = ROOT.TGraph(12, n, Medias)
gr.SetTitle("Tiempo medio vs N")
gr.SetMarkerStyle( 21 )
x=gr.GetXaxis(); y = gr.GetYaxis()
x.SetTitle("N"); y.SetTitle("Tiempo medio [s]")
gr.Fit("sq", "R")
gr.Draw()

c13.Update()
c13.Draw()
```
