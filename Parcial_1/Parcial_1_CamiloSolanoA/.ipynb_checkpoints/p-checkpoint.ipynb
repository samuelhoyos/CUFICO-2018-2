```python
import numpy as np
from scipy.optimize import curve_fit
from scipy import asarray as ar,exp
import matplotlib.pylab as plt
def gaus(x,a,x0,sigma):
        return a*exp(-(x-x0)**2/(2*sigma**2))

file = open("datos.txt", "r") 
fil=file.read().split('\n')[:-1]
data=[]
for f in fil:
    arr=[int(i) for i in f.split(',')[:-1]]
    data.append([x for x in arr if x != max(arr)])

print("4.Maximos pasos realizados:")
print("(Dimension - Pasos)")
for n in range(len(data)):
    print(str(5+n*5)+" - "+str(max(data[n])))
```

```python
print("5.Parametros obtenidos en ajuste de gaussiana:")
print("(Dimension  , Media  ,  Desviacion estandar)")
l_med=[]
for n in range(len(data)):
    y,x=np.histogram(data[n],bins=10)
    x=x[:-1]
    l = len(x)
    mean = np.mean(data[n])
    sigma = np.std(data[n])
    popt,pcov = curve_fit(gaus,x,y,p0=[max(data[n]),mean,sigma])
    
    print("De los datos:",[5+n*5,mean,sigma])
    print("Del ajuste:",[5+n*5,popt[1],popt[2]])
    l_med.append(popt[1])
    plt.plot(x,y,'b+:',label='frecuencia')
    ran=np.linspace(min(data[n]),max(data[n]),30)
    plt.plot(ran,gaus(ran,*popt),'r:',label='ajuste')
    plt.xlabel("Pasos")
    plt.legend()
    
    plt.show()
```

```python
print("6.Funcion que ajusta el numero maximo de pasos:")

x=range(5,61,5)
y=[]
for n in data:
    y.append(max(n))
    
a=np.polyfit(x, y, 2)
print(str(round(a[2],2))+" + "+str(round(a[1],2))+"*x"+" + "+str(round(a[0],2))+"*x^2")
plt.plot(x,y,'b+:',label='valor maximo')
plt.xlabel("Dimension")
ran=np.linspace(min(x),max(x),30)
plt.plot(ran,a[2]+ran*a[1]+ran*ran*a[0],'r:',label='ajuste')
plt.legend()
plt.show()
    
```

```python
print("7.Funcion que ajusta el valor medio de gaussiana:")

x=range(5,61,5)
y=l_med
    
a=np.polyfit(x, y, 2)
print(str(round(a[2],2))+" + "+str(round(a[1],2))+"*x"+" + "+str(round(a[0],2))+"*x^2")
plt.plot(x,y,'b+:',label='valor medio')
plt.xlabel("Dimension")
ran=np.linspace(min(x),max(x),30)
plt.plot(ran,a[2]+ran*a[1]+ran*ran*a[0],'r:',label='ajuste')
plt.legend()
plt.show()
```
