Para cada n se cargan los archivos que contienen los datos, estos se grafican en un histograma al cual se le hace un fit con una gaussiana, con la cual se obtiene el valor medio correspondiente

```
import numpy as np
from ROOT import gROOT, TCanvas, TH1F, TStyle, TLatex
```

```
X = np.loadtxt('drive/My Drive/datos_n5.txt', unpack = True)
histo5 = ROOT.TH1F( 'h1f', 'tiempos de estabilizacion (n=5)', 250, 0, 0.02 )
for j in X:
  histo5.Fill(j)
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillStyle(1)
f1 = ROOT.TF1("f1", "gaus", 0.001, 0.04);
histo5.Fit("f1","R")
histo5.Draw()
x = histo5.GetXaxis()
x.SetTitle("Tiempo [s]")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```
