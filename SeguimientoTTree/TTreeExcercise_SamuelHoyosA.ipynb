```
import ROOT
```

```
from array import array
```

```
MyTree = ROOT.TTree( 'tree1', 'tree para tres variables' )
```

```
x = array( 'd', [ 0 ] )
y = array( 'd', [ 0 ] )
z=array("d", [0])
```

```
MyTree.Branch( 'x', x, 'x/D' )
MyTree.Branch( 'y', y, 'y/D' )
MyTree.Branch("z",z,"z/D" )
```

```
import numpy as np
from  random import uniform 
```

```
for i in range(10000):
  x[0] = np.random.normal(loc=5,scale=2)
  y[0] = np.random.normal(loc=5,scale=2)
  z[0]=uniform(0,10)
  MyTree.Fill()
  
print (y)
```

```
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x*y>>1(28,0,100)")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x*z>>2(28,0,100)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x>>4(14,0,10)","4.5<=z && z<=5.5")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x*y*z>>3(42,0,1000)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x:z>>5(14,0,10,14,0,10)","4.5<=y && y<=5.5", "lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x:y:z>>6(14,0,10,14,0,10,14,0,10)","", "lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```
